## Training & Test Dataset

Sebelum melakukan proses training model, umumnya dataset akan dibagi menjadi 2 bagian yang biasa dikenal dengan istilah **training set** dan **test set**. Proses pembagiannya dilakukan secara acak. Modul Scikit-Learn sudah memfasilitasi kita untuk melakukan proses ini dengan lebih mudah. 

Nah untuk melakukan splitting atau pembagian data set menjadi training dan testing set, disini pertama kita import dulu function `train test split`. Caranya cukup mudah kita panggil saja `sklearn.model_selection` lalu kita panggil `import train_test_split`. 

````python
from sklearn.model_selection import train_test_split

# panggil fungsi train_test_split, butuh 4 parameter, 
X_train, X_test, y_train, y_test = train_test_split(X, # features 
                                                   y, # target
                                                   test_size=0.3, # test size, value dari 0-1, 0.3 mempresentasikan data test 0.3 atau 30%, 0.7 train  
                                                   random_state=1) # acak secara konsisten

print(f"X train: {X_train.shape}")
print(f"X test: {X_test.shape}")
print(f"y train: {y_train.shape}")
print(f"y test: {y_test.shape}")
````

Fungsi dari `train_test_split` ini akan menghasilkan 4 kelompok data, oleh karena itu kita perlu menampungnya ke dalam 4 variable yaitu : `X_train`, `X_test`, `y_train` dan `y_test`
- `X_train` akan menampung **features** untuk **training set**
- `X_test` akan menampung **features** untuk **testing set**
- `y_train` akan menampung **target** untuk **training set**
- dan `y_test` akan menampung **target** untuk test **testing set**

Lalu fungsi .shape akan mencoba untuk menampilkan dimensi dari masing - masing variable seperti `X_train`, `X_test`, `y_train` dan `y_test`

untuk output :

```python
X train: (105, 2)
X test: (45, 2)
y train: (105,)
y test: (45,)
```

- untuk bagian X_train dan y_train sama-sama terdiri dari 105 baris
- dan untuk X_train dan X_test terdiri dari 45 baris
- yang perlu diperhatikan adalah jika kolom di X_train dan y_train hanya terdiri dari 2 baris, jangan bingung. itu terjadi karena sebelum plotting kira sudah menyimpan di variable X jika kolomnya hanya akan kita ambil 2 pada bagian `X = X[:, :2]`
