# 01 Pengenalan Machine Learning

## Apa itu Machine Learning (ML)
- "_`Machine Learning` is a study that gives computers the ability to learn without explicitly programmed_" **- Arthur Samuel**
    > Nah definisi ini di kemukakan oleh seorang _computer scientist_ bernama **Arthur Samuel**. Nah disini Arthur Samuel berpendapat bila Machine Learning itu adalah suatu bidang studi atau bidang keilmuan yang memampukan suatu komputer untuk belajar tanpa secara eksplisit kita program.
- "_`Machine Learning` is a program can be said to learn from experience ***E*** with respect to some class of tasks ***T*** and performence measure ***P***, if its performance at tasks in ***T***, as measured by ***P***, improves with experience ***E***_" **- Tom Michell**
    > Nah difinisi ini di kemukakan oleh _computer scientist_ yang lain bernama **Tom Michell**. Disini Tom berpendapat bahwa yang namanya _Machine Learning_ itu adalah program yang belajar dari pengalaman atau _Experience_ **E** terhadap suatu _task_ atau suatu tugas tertentu dimana performanya bisa diukur sebagai **P**. Dimana nilai performa ini akan terus meningkat seiring dengan peningkatan _experience_ atau atau pengalaman dari si program dalam menyelesaikan tugas **T** tadi.
    
    > Nah, ini merupakan kedua definisi _machine learning_ yang cukup umum ditemui. atau dengan kata lain kalau mau kita sederhanakan dalam bahasa Indonesia kita bisa juga definisikan :
- _`Machine Learning`_ adalah bidang ilmu yang memampukan suatu program komputer untuk belajar dari sekumpulan data.

## Learning Problem 
- **Learning Problem** dalam konteks Machine Learning (ML) melibatkan suatu dataset (terdiri dari sejumlah $n$ sample data) untuk melakukan prediksi terhadap properties yang tidak diketahui pada dataset lain yang sejenis.
- Secara umum learning problem dapat dibagi menjadi dua, yaitu **supervised learning** dan **unsupervised learning**.
- Pada **supervised learning problem**, program memprediksi suatu nilai output untuk suatu input berdasarkan proses **learning** dengan memanfaatkan sekumpulan data yang terbagi dalam dua label, yaitu input dan output. Disini program melakukan **learning** dari sekumpulan contoh kasus (examples) yang menyertakan "jawaban benar"
- Pada **unsupervised learning problem**, program tidak melakukan learning dari labeled data; melainkan mencoba menemukan pola (pattern) pada data. Di sini program melakukan **learning** dari sekumpulan contoh kasus (example) tanpa disertai "jawaban benar". 

---

## Terminologi Dasar pada Supervised Learning
- **output** : label, response variable, dependent variable, regressand, criteron variable, measured variable, responding variable, explained variable, outcome variable, experimental variable, dan output variable.
- **input** : features, explanatory variables, predictors, regressors, cotrolled variables dan exposure variables.
- **training set** : Sekumpulan "contoh" yang digunakan untuk proses **learning (training)** pada supervised learning.
- **test set** : Sekumpulan "contoh" yang digunakan untuk mengukur performa pada supervised learning.
- Pada supervised learning, **response variable (output)** bisa dianalogikan sebagai kunci jawaban dan **explanatory variables (input)** bisa dianalogikan sebagai **pertanyaan**nya. 

---

## Supervised Learning Task
- Secara umum terdapat dua jenis Supervised Learning Machine Learning Task, yaitu **Classification** dan **Regression**.
- **Classification Task** menekankan pada proses learning (training) untuk melakukan prediksi nilai discrete (i.e., category, class, label) dari sejumlah features.
- **Regression Task** menekankan pada proses learning (training) untuk melakukan prediksi nilai countinous (floating point) pada sejumlah features.
> Dengan kata lain, kalau kita ingin sederhanakan Classification task ini akan fokus pada pengelompokan kategori. Sedangkan Regression task akan berfokus pada nilai countinous. 

---

## Unsupervised Machine Learning Task
- Terdapat beberapa Unsupervised Machine Learning Task dan kita akan berfokus pada dua tasks, yaitu Clustering dan Dimensionality Reduction.
- **Clustering Task** menekankan pada proses explorasi data untuk menemukan kelompok (group) pada data berdasarkan kemiripan (kedekatan) karakteristik.
- **Dimensionality Reduction Task** menekankan pada proses explorasi data untuk menemukan sejumlah features yang paling berdampak pada response variable.

## Rekomendasi Buku
- **Machine Learning Yearning** (by Andrew Ng)
> Buku ini termasuk buku yang sangat baik sekali untuk memberikan pengantar konseptual terkait Machine Learning. Bukunya juga bisa diperoleh dengan gratis. Buku Machine Learning ini juga termasuk ramah bagi pemula. Dan setiap konsep Machine Learning yang disampaikan benar-benar dikemas dengan sangat sederhana sekali.
- **The Hundred-Page Machine Learning Book** (by Andriy Burkov)
> Nah buku ini juga cukup menarik dan cukup padat, hanya saja penekanannya sedikit berbeda dari buku yang pertama. Buku kedua ini mencoba untuk membahas beberapa model yang memang umum dan populer dalam bidang machine learning. Nah, karena penekanannya disini adalah pembahasan terhadap beberapa model machine learning, maka buku ini juga sangat syarat akan notasi matematika. Oleh karenannya, untuk dapat memahami buku ini dengan baik, setidaknya kalian membutuhkan dasar matematika yang baik pula. Buku ini masuk dalam rekomendasi mampu membahas setiap model learning yang populer dengan sederhana dan to the point. 
- **Learning from Data** (by Yaser S. Abu-Mustofa)
> Nah buku ini bisa dibilang sebagai text book terbaik bagi teman-teman yang yang memang sedang menjalani perkuliahan atau mengikuti perkuliahan terkait machine learning. Pemaparan konsep machine learning yang disampaikan buku ini benar-benar sangat step by step dan sangat bertahap sekali. Buku ini juga banyak diadopsi oleh banyak universitas di dunia sewaktu mereka menyelenggarakan perkuliahan machine learning.
